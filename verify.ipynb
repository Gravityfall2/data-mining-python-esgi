```python
import pandas as pd

# 1. Charger les deux CSV
df1 = pd.read_csv("attack_detected.csv")
df2 = pd.read_csv("test_prediction.csv")

# 2. Extraire les colonnes (remplace par le vrai nom si besoin)
col1 = df1.iloc[:, 0]  # première colonne du fichier 1
col2 = df2.iloc[:, 0]  # première colonne du fichier 2

# 3. Comparer
resultats = pd.DataFrame({
    "attack_detected": col1,
    "test_prediction": col2,
    "identique": (col1 == col2).astype(int)  # 1 = identique, 0 = différent
})

# 4. Résumé
total = len(resultats)
identiques = resultats["identique"].sum()
differents = total - identiques

print(f"✅ Identiques : {identiques}/{total} ({identiques/total:.1%})")
print(f"❌ Différents : {differents}/{total} ({differents/total:.1%})")

# 5. Afficher les lignes différentes
print("\nLignes différentes :")
display(resultats[resultats["identique"] == 0])

# 6. Exporter le résultat
resultats.to_csv("comparaison.csv", index=False)
```
