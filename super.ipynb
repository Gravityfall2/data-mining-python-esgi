```python
import joblib
import pandas as pd

# 1. Charger le modèle
model = joblib.load('modele_detection_attaques.pkl')

# 2. Préparer les nouvelles données (exemple)
df = pd.read_csv("test_clean.csv")

# 3. Faire des prédictions
predictions = model.predict(df)

# 4. Afficher les résultats
print("Prédictions (0 = pas d'attaque, 1 = attaque) :")

df["prediction"] = predictions

df.head(20)

df[["prediction"]].to_csv("test_prediction.csv", index=False)
```
